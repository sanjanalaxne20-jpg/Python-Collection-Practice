```python
# Create an empty tuple
t2=()
# Print the data type of the tuple
print(type(t2))
```

```python
#create a tuple with interger value
t3=(10,20,30,40,40,50,60)
t3
```

```python
#create a tuple using different data
t4=(1,'ram',78.56)
t4
```

```python
#tuple object does not support insertion operation
t3[3]=50
t3
```

```python
#tuple object does not support item deletion
del t3[3]
```

```python
#print value using index function
t3[1:5:3]#start1 and step 3 and end 5
```

```python
#counting the value in the tuple
t3.count(40)
```

```python
#finding the position of number using index function
t3.index(40)
```

```python
#sorting the tuple in ascending order
sorted(t3)# output return in list , because tuple is immutable( does not changed or iterate the value)
```

```python
#finding the length of tuple
len(t3)
```

```python
#access the element at index 4
t3[4]
```

```python
# Access the last element using negative indexing
t3[-1]
```

```python
t3 = (3,2,9,1,6,8)
sorted_tuple = tuple(sorted(t3))
print(sorted_tuple)
print(type(sorted_tuple))
```
