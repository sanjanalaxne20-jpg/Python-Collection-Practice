```python
#print the type of collection
d2=dict()
print(type(d2))
```

```python
#Accesses the value associated with key 2 from the dictionary.
d3={1:"Ram",2:"Sham",3:"Jay",4:"Vijay"}
d3[2]#output: 'Sham'
```

```python
#replace the key value
d3[4]="om"
d3
```

```python
#adding the key value
d3[5]="Ajay"
d3
```

```python
d3.update({6:"Sachin"})#update the key value
d3
```

```python
## Removes and returns the last key-value pair from the dictionary.
d3.popitem()
d3
```

```python
d3.pop(4)#removes key-value pair from the dictionary at index 4
d3
```

```python
d3.keys()#print only keys
d3
```

```python
d3.values()#print only values
d3
```

```python
d3.items()#print the items
```

```python
del d3[2]#delete the key value pair
d3
```

```python
d3.clear()#clear the dictionary
d3
```
