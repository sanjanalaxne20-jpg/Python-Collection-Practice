```python
#Returns common elements.
A={1,2,3,4,5}
B={4,5,6,7,8}
A.intersection(B)
```

```python
#Returns all unique elements from both sets.
A={1,2,3,4,5}
B={4,5,6,7,8}
A.union(B)
```

```python
#Returns elements present only in the first set.
a = {1, 2, 3}
b = {2, 3, 4}
print(a.difference(b))
```

```python
#Returns elements present in either set but not both.
a = {1, 2, 3}
b = {2, 3, 4}
print(a.symmetric_difference(b))
```

```python
#Updates the first set with common elements.
a = {1, 2, 3}
b = {2, 3, 4}
a.intersection_update(b)
print(a)
```

```python
#Removes common elements from the first set.
a = {1, 2, 3}
b = {2, 3}
a.difference_update(b)
print(a)
```

```python
#Updates the set with non-common elements.
a = {1, 2, 3}
b = {2, 3, 4}
a.symmetric_difference_update(b)
print(a)
```

```python
#Checks whether all elements of the first set are in the second set.
a = {1, 2}
b = {1, 2, 3}
print(a.issubset(b))
```

```python
#Checks whether the first set contains all elements of the second set.
a = {1, 2, 3}
b = {1, 2}
print(a.issuperset(b))
```

```python
#Returns True if the sets have no common elements.
a = {1, 2}
b = {3, 4}
print(a.isdisjoint(b))
```
